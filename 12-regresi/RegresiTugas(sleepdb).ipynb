```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from scipy import stats
```

```python
data = pd.read_excel("sleep_health_data.xlsx")
data
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
len(x_train)
```

```python
print(y_train)
```

```python
data.shape
```

```python
dataku = pd.DataFrame(data)
dataku.head
```

```python
#Lakukan fitting simple linear regression train set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
#Visualisasi Grafik
plt.scatter(dataku.QualityofSleep, dataku.StressLevel)
plt.xlabel('Quality of Sleep')
plt.ylabel('Stress Level')
plt.title('Grafik Kualitas tidur terhadap Stress Level')
plt.show
```

```python
#Visualisasi hasil pred pd trainset
plt.figure(figsize=(10,8)) #untuk uk plotnya

#biru as data observasi
plt.scatter(x_train, y_train, color = 'blue')

#garis merah as hasil pred dari ML
plt.plot(x_train, regressor.predict(x_train), color = 'red')

plt.xlabel('Quality of Sleep')
plt.ylabel('Stress Level')
plt.title('Grafik Kualitas tidur terhadap Stress Level(pada Train Set)')
plt.show
```

```python
#Visualisasi hasil pred pd test set
plt.figure(figsize=(10,8)) #untuk uk plotnya

#biru as data observasi
plt.scatter(x_test, y_test, color = 'blue')

#garis merah as hasil pred dari ML
plt.plot(x_test, regressor.predict(x_test), color = 'red')

plt.xlabel('Quality of Sleep')
plt.ylabel('Stress Level')
plt.title('Grafik Kualitas tidur terhadap Stress Level(pada Test Set)')
plt.show
```
